Task 5: Final Bill Generation and Output Analysis

```python
def calculate_energy_charges(units, customer_type):
    rates = {
        "domestic": 2.5,
        "commercial": 4.0,
        "industrial": 6.0
    }
    return units * rates.get(customer_type, 0)


def calculate_fixed_charges(customer_type):
    charges = {
        "domestic": 50,
        "commercial": 100,
        "industrial": 200
    }
    return charges.get(customer_type, 0)


def calculate_electricity_duty(ec):
    return ec * 0.05


# Input
pu = int(input("Enter Previous Units: "))
cu = int(input("Enter Current Units: "))
ctype = input("Enter Customer Type: ").lower()

units = cu - pu
ec = calculate_energy_charges(units, ctype)
fc = calculate_fixed_charges(ctype)
cc = 30
ed = calculate_electricity_duty(ec)

total_bill = ec + fc + cc + ed

# Output
print("\n----- TGNPDCL ELECTRICITY BILL -----")
print("Units Consumed:", units)
print("Energy Charges (EC): ₹", ec)
print("Fixed Charges (FC): ₹", fc)
print("Customer Charges (CC): ₹", cc)
print("Electricity Duty (ED): ₹", ed)
print("-----------------------------------")
print("Total Bill Amount: ₹", total_bill)
```
